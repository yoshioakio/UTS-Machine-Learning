library pandas = digunakan untuk membuat dataframe dari data fakultas, jumlah mahasiswa, dan akreditasi.
library matplotlib.pyplot = digunakan untuk membuat plot bar yang menampilkan jumlah mahasiswa per fakultas.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

membuat data struktural yang dapat divisualisasikan dalam bentuk plot bar(blok), untuk  menampilkan distribusi jumlah mahasiswa per fakultas dan tingkat akreditasi fakultas berdasarkan data yang dibuat.

```python
fakultas = ["Bisnis", "D3 Perhotelan", "ICT", "Ilmu Komunikasi", "Seni dan Desain"]
jumlah_mahasiswa = [260, 28, 284, 465, 735]
akreditasi = ["A","A","B","A","A"]

info_mahasiswa = pd.DataFrame({
    "fakultas": fakultas,
    "jumlah_mahasiswa": jumlah_mahasiswa,
    "akreditasi": akreditasi
})
```

Membuat code untuk menampilkan hasil outputnnya

```python
print(info_mahasiswa)
plt.figure(figsize=(6, 3))
plt.bar(info_mahasiswa["fakultas"], info_mahasiswa["jumlah_mahasiswa"], color=['red', 'orange', 'green', 'blue', 'purple'])
plt.xlabel('Fakultas')
plt.ylabel('Jumlah Mahasiswa')
plt.title('Jumlah Mahasiswa per Fakultas')
plt.tight_layout()
plt.show()
```
